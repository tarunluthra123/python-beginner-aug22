### Write a function which takes one number as argument and prints its square.

````python
def square(x):
    # Check if type of x is int or float
    if isinstance(x, (int, float)):
        print(x * x)
    else:
        print("Invalid input")
````

````python
square(5)
````

````python
square(9)
````

````python
square("Hello")
````

````python
square(5.6)
````
